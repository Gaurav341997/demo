# Now applying base model as random forest 

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
x=df.drop(columns=["Close"])
y=df[['Close']]
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=33,test_size=0.2)
```

```python
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(train_x,train_y)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
pred=rf.predict(test_x)
```

```python
r2_score(test_y,pred)
```

```python
mean_absolute_error(test_y,pred)
```

```python
mean_squared_error(test_y,pred)
```

```python
np.sqrt(mean_squared_error(test_y,pred))
```

```python
rf2=RandomForestRegressor(n_estimators=150,random_state=44)
```

```python
rf2.fit(train_x,train_y)
```

```python
pred=rf2.predict(test_x)
pred_tr=rf2.predict(train_x)
```

```python
r2_score(test_y,pred)
```

```python
mean_absolute_error(test_y,pred)
```

```python
r2_score(train_y,pred_tr)
```
